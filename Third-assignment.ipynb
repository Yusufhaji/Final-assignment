# Question: Do countries with low initial levels of capital grow faster than and catch up with countries that start off with high levels of capital?

 ## Motivation
 
 
We will simulate the Unconditional (or absolute) convergence and find out whether 
all countries converge to the same steady-state level of per capita income and whether this is irrespective of the level
of savings(s) and population growth rate(n). We will simulate a simple Solow model and find out whether the suggested implication of the Solow model that countries with low initial levels of output and capital will grow faster than and catch
up with countries that start off with high levels of output will hold in the long run.

In this simulation we are trying to identify how fast a country will converge to a steady state equilibrium k*, depending on the initial level of investment and savings. The motivation for this simulation is the “Absolute convergence hypothesis”: In the neoclassical model of exogenous growth, the economy tends to grow faster for a low level of initial capital-labor ratio compared to the case of a higher capita-labor ratio. This could be otherwise stated as when countries have similar characteristics, the ‘poor’ countries with low initial income y0, will display a higher growth rate and the two variables are expected to be negatively correlated.
Assumptions of the Basic Solow Model
* Aggregate Production Function Y = A KaL1-a or in per capita terms y = AKa
* Given constant rate of technology A
* Labor Supply growing at rate n
* Exogenous Saving Rate
* Constant depreciation rate δ
* Inada Conditions
Saving Rates in the Solow Model
Solow model predicts that poor countries will converge towards income levels of rich countries, if they have relatively small saving rates. This happens because in the Solow model (both with or without exogenous technological progress) an increase in the saving rate raises growth in the short run, but not in the long run, once the economy has again reached the new steady state. Intuitively, the relationship between saving rate and convergence to a steady state k* is negative, implying that the smaller the saving rate, the higher the level of investment on capital, the faster the convergence towards the steady state k*.
An economy will transition towards a higher level of steady state if there is an increase in its rate of saving or a decrease in the rate of population growth. However, it will converge faster towards a steady state, the lower the saving rate. As long as this holds, and, from the assumptions, the only exogenous variable is the saving rate, we want to see the different convergence equilibria that we will get by manipulating the different values of the saving rate. In the context of this simulation, we have an ‘absolute convergence hypothesis’, in the sense that we assume all the structural characteristics of the countries to be identical (n,g,δ) except the value of the saving rate, which we change to see the different outcomes in the steady state k*.




 

## Method

The Solow model takes the rates of savings, population growth and technological progress as exogenous.
A Cobb-Douglas production function is assumed , so production at time t is given by;

$Y(t) = K(t)^α (A(t)L(t))^{1-α}  ,0<α<1$

The Solow growth model is a neoclassical growth model where the amount of capital stock per capita k_t evolves according to the rule:

$k_(t+1)=(szk_(t+1)+(1-δ)k_t)/(1+n)$

Whereby:
* s is an exogenously given savings rate
* z is a productivity parameter
* α is capital’s share of income
* n is the population growth rate
* δ is the depreciation rate

 The steady state of the Solow model is the k that solves equation(1)  when 
 
 $k_(t+1)=k_t=k$    


L and A are assumed to be exogenous at rates n and g∶ $L(t) = L(0)  e^{nt}$

$A(t)= A(0)  e^{gt}$

Defining variables in intensive form: y ≡ Y/AL  , k ≡ K/AL  

Therefore ;

$y = K^α (AL)^{1-α} = k^α = (K/AL)^{α}$

$y = k^α$

Capital accumulates and the evolution of capital is governed by:

$k ̇(t) = sy(t)- (n+δ)k(t)$

$= k ̇(t) = sy(t)- (n+δ)k(t)$

And so the steady state capital per effective worker is :

We will therefore simulate the below equation to see how different countries with different initial savings rates converge in the long run. Also we simulate and see how countries with same initial savings rate converge in the long run.  We will consider the below model: 

$k_(t+1)=(szk_(t+1)+(1-δ)k_t)/(1+n)$


We will therefore simulate in order confirm the existence of unconditional convergence. We will also check the existence of conditional convergence.




```python
"""
Filename: solow.py
Reference: http://quant-econ.net/py/python_oop.html
"""
from __future__ import division  
import numpy as np

class Solow:
    r"""
    Implements the Solow growth model with update rule

    .. math::
        k_{t+1} = \frac{s z k^{\alpha}_t}{1 + n}  + k_t \frac{1 + d}{1 + n}
    
    """

    def __init__(self, n, s, d, alpha, z, k):
        """ 
        The Solow model with Cobb-Douglas production function.  
        """
        self.n, self.s, self.d, self.alpha, self.z = n, s, d, alpha, z
        self.k = k
        

    def h(self,x):
        
        temp = self.s * self.z * self.k**self.alpha + self.k * (1 - self.d)
        return temp / (1 + self.n)

    def update(self):
       
        self.k =  self.h(self.k)
        
    def steady_state(self):
         
         return ((self.s * self.z) / (self.n + self.d))**(1 / (1 - self.alpha))
     
    def generate_sequence(self, t):
        
        path = []
        for i in range(t):
            path.append(self.k)
            self.update()
        return path
```

```python
import matplotlib.pyplot as plt
baseline_params = 0.05, 0.25, 0.1, 0.3, 2.0, 1.0
s1 = Solow(*baseline_params) 
s2 = Solow(*baseline_params)
s2.k = 8 # This is for the high economy
T = 60
fig, ax = plt.subplots()
# We plot the common steady-state level of capital
ax.plot([s1.steady_state()]*T, 'k-', label='steady state')

for s in s1, s2:
    lb = 'capital from the  initial savings rate {}'.format(s.k)
    ax.plot(s.generate_sequence(T), 'o-', lw=2, alpha=0.6, label=lb)

ax.legend(loc='lower right')
plt.show()
```

## Conclusion

The Solow model allows economists to view capital accumulation and the technological progress overtone in a framework which is easy to track. If a country is not present in the environment of Ak, then no technological progress would exist according to the model 
The model illustrates steady state levels output and capital within the parameters. When savings rates increase, the steady state output increased, causing the economy to grow into a new steady state. When this point is reached growth will decline and eventually stop. Therefore the Solow model illustrates that a factor, witch an increase over time, causes the growth. An increase in technology or population growth are one of these factors. Capital accumulation is determined by the savings rate, population growth and the rate of depreciation. Output growth, per capita is exogenously determined by technological progress.

When applied to the targeted county in the real world, a number of things are bound to happen. One if these events takes place if growth due to capital accumulation occurs, growth would be very strong when countries first accumulate capital. This growth will eventually slow down as shown similar in our graphs. This applied to for example Asian countries in the 50’s and 60’s. Counties like these also converge in output per capita. This will lead counties like these to catch up to developed countries where all countries have reached a steady state, assuming the production function is similar. However these assumptions do not always hold as there are exemptions in the model as seen throughout history. Less developed Asian countries are actually seen to fall behind developing countries, unlike the model’s prediction which suggests that they will catch up. Another example can be found in the Sovjet Union under Josef Stalin. Compared to the United States, the Sovjet Union had higher saving due to its lower level of capital. This would suggest that the Sovjet Union would catch up, according to the model. It did not. These event might be explained by other factors such as productivity increase, but it illustrates the limitations of the Solow model.
